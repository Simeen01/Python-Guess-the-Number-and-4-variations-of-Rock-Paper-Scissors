**Guess the number**

````python
import random
computer = random.randint(1,100)
# print(computer)
i=1
chances = 7

while(i<=chances):
  user = int(input("\nGuess the Number: "))
  if i!=7:
    print("You have",chances-i,"chances left.")
  
  if i>=chances and (user>computer or computer>user):
    print("Game Over!")  
  elif user>computer:
    print("Your number is greater than the actual number.\nPlease enter a smaller number.")
  elif computer>user:
    print("Your number is smaller than the actual number.\nPlease enter a larger number.")
  else:
    print("Congratulations!! You guessed the number right.")
    print("You have won in",i,"chances only!")
    break
  i+=1
````

*Rock - Paper - Scissor with 2 players*



```
*   Invalid inputs - revert back to input
*   Hide input using getpass()
*   User chooses whom to play with - computer or 2 player
```



````python
import getpass

play = ["Rock","Paper","Scissor"]
user1_wins = 0
user2_wins = 0
user1_name = input("Player1, please enter your name: ")
user2_name = input("Player2, please enter your name: ")
x = 0

#no. of rounds
rounds = int(input("How many rounds do you want to play? "))
i=0
while(i<rounds):
  if x==0:
    user1 = getpass.getpass(prompt=user1_name+", What would you like (Rock/Paper/Scissor)? ")
    user1 = user1.capitalize()
    if user1 not in play:
      print("Invalid input. Please try again!")
      continue

  
  user2 = getpass.getpass(prompt=user2_name+", What would you like (Rock/Paper/Scissor)? ")
  user2 = user2.capitalize()
  if user2 not in play:
    print("Invalid input. Please try again!")
    x = x + 1
    continue
  
  print("\nInputs:-")
  print(user1_name,": ",user1)
  print(user2_name,": ",user2)

  #rock wins over scissor
  #if user chooses rock
  if user1 == "Rock":
    if user2 == "Scissor":
      print(user1_name,"Wins!")
      user1_wins +=1
    elif user2 == "Paper":
      print(user2_name,"Wins!")
      user2_wins +=1
    else:
      print("Draw!")

  #if user chooses paper
  #paper wins over rock
  if user1 == "Paper":
    if user2 == "Rock":
      print(user1_name,"Wins!")
      user1_wins +=1
    elif user2 == "Scissor":
      print(user2_name,"Wins!")
      user2_wins +=1
    else:
      print("Draw!")

  #if user chooses Scissor
  #paper wins over rock
  if user1 == "Scissor":
    if user2 == "Paper":
      print(user1_name,"Wins!")
      user1_wins +=1
    elif user2 == "Rock":
      print(user2_name,"Wins!")
      user2_wins +=1
    else:
      print("Draw!")
  i+=1
  x = 0

#score
print("\nScore")
print("-----")
print(user1_name,":",user1_wins)
print(user2_name,":", user2_wins)

#winner annoncement
if user1_wins>user2_wins:
  print("Congratulations!",user1_name,"wins the game.")
elif user2_wins>user1_wins:
  print("Congratulations!",user2_name,"wins the game.")
else:
  print("The game was a Draw!")
````

````python
import random
import getpass

play = ["Rock","Paper","Scissor"]
user1_wins = 0
user2_wins = 0
computer_wins = 0
user_wins = 0
x = 0 #for allowing only user2 to re-enter response, if user1 has given right response
i=0

print("Rock-Paper-Scissor Game")
print("-----------------------")
# choices = 0
while True:
  print("\nChoose who you want to play with: ")
  print("1. Computer")
  print("2. 2 Players")
  print("3. Exit")

  player_dict = {
    1: "1 player",
    2: "2 player"
  }
  option = int(input("\nChoose your option: (1/2/3)\n"))

  #1 player : user - computer
  if option == 1 in player_dict:
    name = input("Enter your name: ")
    #no. of rounds
    rounds = int(input("How many rounds do you want to play?"))

    while(i<rounds):
      user = input(name+", What would you like (Rock/Paper/Scissor)? ")
      user = user.capitalize()
      if user not in play:
        print("Invalid input. Please try again!")
        continue
      computer = random.choice(play)
      
      print("\nInputs:-\nUser: ",user)
      print("Computer: ",computer)

      #rock wins over scissor
      #if user chooses rock
      if user == "Rock":
        if computer == "Scissor":
          print("You Win!")
          user_wins +=1
        elif computer == "Paper":
          print("Computer Wins!")
          computer_wins +=1
        else:
          print("Draw!")

      #if user chooses paper
      #paper wins over rock
      if user == "Paper":
        if computer == "Rock":
          print("You Win!")
          user_wins +=1
        elif computer == "Scissor":
          print("Computer Wins!")
          computer_wins +=1
        else:
          print("Draw!")

      #if user chooses Scissor
      #paper wins over rock
      if user == "Scissor":
        if computer == "Paper":
          print("You Win!")
          user_wins +=1
        elif computer == "Rock":
          print("Computer Wins!")
          computer_wins +=1
        else:
          print("Draw!")
      i+=1

    #score
    print("\nScore")
    print("-----")
    print("Your score:",user_wins)
    print("Computer's score:", computer_wins)

    #winner annoncement
    if user_wins>computer_wins:
      print("Congratulations!",name,"wins the game.")
    elif computer_wins>user_wins:
      print("Computer wins the game.")
    else:
      print("The game was a Draw!")


  #2 player : user1 - user2
  elif option == 2:
    user1_name = input("Player1, please enter your name: ")
    user2_name = input("Player2, please enter your name: ")
    #no. of rounds
    rounds = int(input("How many rounds do you want to play?"))

    while(i<rounds):
      if x==0:
        user1 = getpass.getpass(prompt=user1_name+", What would you like (Rock/Paper/Scissor)? ")
        user1 = user1.capitalize()
        if user1 not in play:
          print("Invalid input. Please try again!")
          continue

      
      user2 = getpass.getpass(prompt=user2_name+", What would you like (Rock/Paper/Scissor)? ")
      user2 = user2.capitalize()
      if user2 not in play:
        print("Invalid input. Please try again!")
        x = x + 1
        continue
      
      print("\nInputs:-")
      print(user1_name,": ",user1)
      print(user2_name,": ",user2)

      #rock wins over scissor
      #if user chooses rock
      if user1 == "Rock":
        if user2 == "Scissor":
          print(user1_name,"Wins!")
          user1_wins +=1
        elif user2 == "Paper":
          print(user2_name,"Wins!")
          user2_wins +=1
        else:
          print("Draw!")

      #if user chooses paper
      #paper wins over rock
      if user1 == "Paper":
        if user2 == "Rock":
          print(user1_name,"Wins!")
          user1_wins +=1
        elif user2 == "Scissor":
          print(user2_name,"Wins!")
          user2_wins +=1
        else:
          print("Draw!")

      #if user chooses Scissor
      #paper wins over rock
      if user1 == "Scissor":
        if user2 == "Paper":
          print(user1_name,"Wins!")
          user1_wins +=1
        elif user2 == "Rock":
          print(user2_name,"Wins!")
          user2_wins +=1
        else:
          print("Draw!")
      i+=1
      x = 0

    #score
    print("\nScore")
    print("-----")
    print(user1_name,":",user1_wins)
    print(user2_name,":", user2_wins)

    #winner annoncement
    if user1_wins>user2_wins:
      print("Congratulations!",user1_name,"wins the game.")
    elif user2_wins>user1_wins:
      print("Congratulations!",user2_name,"wins the game.")
    else:
      print("The game was a Draw!")
  
  elif option == 3:
    print("Thank you for playing!") 
    break

  #Invalid choice for choosing whom to play with    
  else:
    print("Invalid input")
  # choices+=1
  continue
````
